```python
import pandas as pd

df = pd.read_csv('Lol_ProfessionalGames.csv')

pd.set_option('display.max_columns', None)
df.columns = df.columns.str.replace(" ", "")
df
```

```python
cuenta = 0
for col in df.columns: 
    if 'azul' in col: 
        cuenta += 1
    if 'rojo' in col: 
        cuenta += 1
print(cuenta)
```

Hay 508 variables que hacen referencia a si son del equipo rojo o el azul por partida, las 9 variables restantes son variables que al convertir el dataset en una fila por equipo de partida se duplican y aparecen en las dos. 

```python
azules = []
rojas = []
globales = []
for col in df.columns: 
    if 'azul' in col: 
        azules.append(col)
    elif 'rojo' in col: 
        rojas.append(col)
    else: globales.append(col)
print(globales)
```

Realmente vemos que las 6 primeras variables son como hemos comentado, pero las 2 de primera sangre, y la primera torre realmente son también de un equipo, habrá que convertirlas para que si el equipo la ha conseguido se indique como propia y no por color. Primero subdivido el dataset en equipos rojos y azules. 

```python
azules = globales + azules
rojas = globales + rojas
```

```python
import numpy as np
azules_df = df[azules]
azules_df['primera_sangre'] = np.where(azules_df['primera_sangre'] == azules_df['nombre_azul'], 1, 0)
azules_df['primera_torre'] = np.where(azules_df['primera_torre'] == azules_df['nombre_azul'], 1, 0)
primera_sangre_jugador = []
for valor in azules_df['First_Blood']: 
    if type(valor) != str:
        primera_sangre_jugador.append(valor)
    else:
        if 'rojo' in valor: 
            primera_sangre_jugador.append(valor.replace('rojo','enemigo'))
        if 'azul' in valor: 
            primera_sangre_jugador.append(valor.replace('azul','equipo'))

azules_df['First_Blood'] = primera_sangre_jugador
azules_df.columns = azules_df.columns.str.replace("_azul", "")
azules_df['lado'] = 'azul'
```

```python
rojas_df = df[rojas]
rojas_df['primera_sangre'] = np.where(rojas_df['primera_sangre'] == rojas_df['nombre_rojo'], 1, 0)
rojas_df['primera_torre'] = np.where(rojas_df['primera_torre'] == rojas_df['nombre_rojo'], 1, 0)
primera_sangre_jugador = []
for valor in rojas_df['First_Blood']: 
    if type(valor) != str:
        primera_sangre_jugador.append(valor)
    else:
        if 'rojo' in valor: 
            primera_sangre_jugador.append(valor.replace('rojo','equipo'))
        if 'azul' in valor: 
            primera_sangre_jugador.append(valor.replace('azul','enemigo'))

rojas_df['First_Blood'] = primera_sangre_jugador
rojas_df.columns = rojas_df.columns.str.replace("_rojo", "")
rojas_df['lado'] = 'rojo'
```

```python
equipos_df = pd.concat([azules_df,rojas_df])
equipos_df
```
